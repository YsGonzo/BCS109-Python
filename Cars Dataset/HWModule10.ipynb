```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import missingno as msno

df = pd.read_csv("https://raw.githubusercontent.com/YsGonzo/Colab/main/Cars%20Dataset/cars.csv")
df.head()
```

```python
df.describe()
```

##Statistics

```python
#Checks for null values
df.isnull().sum()
```

```python
#Checks how many unique values exist
df.nunique()
```

```python
#Shows car counts at different gas mileages.
sns.catplot(x='Fuel Information.Highway mpg', data=df, kind='count', height = 3, aspect = 5)
```

```python
#Visualizes the engine horsepower's relationship with highway mileage.

sns.scatterplot(x='Engine Information.Engine Statistics.Horsepower', y='Fuel Information.Highway mpg', data=df)
plt.ylim(10, 45)
plt.xlim(80, 675)
```

```python
#Visualizes the engine Torque's relationship with highway mileage.
sns.scatterplot(x='Engine Information.Engine Statistics.Torque', y='Fuel Information.Highway mpg', data=df)
plt.ylim(10, 45)
```

```python
#Visualizes the gas mileage for different fuel types in car engines.
df.boxplot(by = 'Fuel Information.Fuel Type', column = ['Fuel Information.Highway mpg'], grid = False)
plt.ylim(10, 45)
```

```python
#Visualization of gas mileage for cars of each Brand.
plt.figure(figsize=(20,20))
plt.barh('Identification.Make', 'Fuel Information.Highway mpg', data =df)
plt.show()
```

#What is Data Science
Data science is about using data to create the most impact possible. Solving realistic problems with data.

Data science began in 2001 when computer science and data mining were put together. Eventually companies developed large unstructured datasets that data scientist undertook, leading to a newfound machine learning approach (This is not the only branch of data science but it became the center of media attention over the years). Key aspects of Data science are shown to be analytics, metrics, a/b testing, and experimentation which are prioritized in a company focused on profits or efficiency, etc. Different sized companies hold different approaches, for example a smaller company would most likely not be worried about AI or deep learning until it is sufficiently large enough.
